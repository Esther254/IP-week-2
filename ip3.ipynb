```python
#finding outliers 
#import required libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df.boxplot(column=["against_fire"],by="against_bug",grid=False)
```

```python
#dealing with outliers
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
df_out=df[~((df<(Q1-1.5*IQR)) | (df>(Q3+1.5*IQR))).any(axis=1)]
#printing out the shape of our new dataset
print(df_out.shape)
#printing out the shape of our old dataset
print(df.shape)
```

```python
#dealing with duplicated values
df.drop_duplicates
```

```python
# fitting label encoder method in the column of the data
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df.info()
```

```python
label_encoder.fit(df["type1"])
```

```python
#view labels in the column
list(label_encoder.classes_)
#transform categories into integers
df["type1"]=label_encoder.transform(df["type1"])
df
df.info()
```
